```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

#1. Feature Engineering

# a. Total Pixel Area
df['total_pixels'] = df['px_height'] * df['px_width']

# b. Screen Area
df['screen_area'] = df['sc_h'] * df['sc_w']

# c. Camera Strength
df['total_cam'] = df['fc'] + df['pc']

# d. Weight to Battery Ratio
df['weight_battery_ratio'] = df['mobile_wt'] / df['battery_power']

# e. Memory-Dependency Interaction
df['memory_dependency'] = df['int_memory'] * df['m_dep']

# 2. Feature Scaling

numeric_cols = ['battery_power', 'clock_speed', 'fc', 'int_memory', 'm_dep',
                'mobile_wt', 'n_cores', 'pc', 'px_height', 'px_width', 'ram',
                'sc_h', 'sc_w', 'talk_time', 'total_pixels', 'screen_area',
                'total_cam', 'weight_battery_ratio', 'memory_dependency']

scaler = StandardScaler()

# Fit and transform the numeric columns
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# 3. Handling Categorical/Binary Features

# List of binary columns
binary_cols = ['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi']

df[binary_cols] = df[binary_cols].astype(int)

# 4. Splitting Features and Target

target = 'price_range'

# Separate features (X) and target (y)
x = df.drop(columns=[target])
y = df[target]


```

```python
# 5. Train-Test Split

from sklearn.model_selection import train_test_split

# Split the data into training and testing sets (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)

# Now, x_train and x_test are ready for model training and evaluation

print(f'Training Set : {x_train.shape},Testing Set : {x_test.shape}')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

model = LogisticRegression()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test,y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f'Accuracy : {accuracy}')
print(f'Precision : {precision}')
print(f'Recall : {recall}')
print(f'F1-Score : {f1}')
print('Confusion Matrix : \n',confusion_matrix(y_test,y_pred))

# import joblib
# joblib.dump(model,'mobile_price_prediction_model_logistic_regression_model.pkl')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

model = RandomForestClassifier()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# import joblib
# joblib.dump(model,'mobile_price_prediction_model_RandomForest.pkl')
```
